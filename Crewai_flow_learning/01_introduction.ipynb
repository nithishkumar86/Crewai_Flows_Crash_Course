```python
from crewai.flow.flow import Flow, start, listen

class GreetingFlow(Flow):

    @start()
    def start_flow(self):
        print("Flow started")
        return "Nithish"

    @listen(start_flow)
    def greet(self, name):
        print(f"Hello {name}, welcome to CrewAI Flows!")
```

```python
greet = GreetingFlow()
```

```python
result = greet.kickoff()
```
